## hypothesis 

```python
import pandas as pd
from scipy.stats import ttest_ind
data=df
# Load data from the CSV file into a DataFrame
# Replace 'data.csv' with your actual file path
data = pd.read_csv('heart.csv')

# Splitting the data into two groups based on 'HeartDisease' status
with_disease = data[data['HeartDisease'] == 1]['Age']
without_disease = data[data['HeartDisease'] == 0]['Age']

# Performing independent t-test for means of the two groups
t_statistic, p_value = ttest_ind(with_disease, without_disease, nan_policy='omit')

# Displaying the results
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Interpretation of results
alpha = 0.05  # significance level
if p_value < alpha:
    print("Reject null hypothesis: There is a significant difference in mean age between individuals with and without heart disease.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in mean age between the groups.")
```
